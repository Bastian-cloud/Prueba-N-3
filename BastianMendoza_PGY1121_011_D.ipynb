```python
class Asistente:
    def __init__(self, run, tipo_entrada):
        self.run = run
        self.tipo_entrada = tipo_entrada


class EntradaConcierto:
    def __init__(self, nombre, precio, cantidad_asientos, rango_asientos):
        self.nombre = nombre
        self.precio = precio
        self.cantidad_asientos = cantidad_asientos
        self.rango_asientos = rango_asientos
        self.asientos_vendidos = []

    def asientos_disponibles(self):
        return list(set(self.rango_asientos) - set(self.asientos_vendidos))

    def vender_asiento(self, asiento, run):
        if asiento in self.asientos_vendidos:
            print("Este asiento no está disponible.")
            return False
        if len(self.asientos_vendidos) >= self.cantidad_asientos:
            print("No quedan más asientos disponibles.")
            return False
        self.asientos_vendidos.append(asiento)
        asistente = Asistente(run, self.nombre)
        asistentes.append(asistente)
        return True


def mostrar_menu():
    print("======= CONCIERTO VIP DE MICHAEL JAM =======")
    print("1. Comprar entradas.")
    print("2. Mostrar ubicaciones disponibles.")
    print("3. Ver listado de asistentes.")
    print("4. Mostrar ganancias totales.")
    print("5. Salir.")


def comprar_entradas():
    tipo_entrada = seleccionar_tipo_entrada()
    if tipo_entrada is None:
        return

    cantidad_entradas = int(input("Ingrese la cantidad de entradas a comprar (máximo 3): "))
    if cantidad_entradas < 1 or cantidad_entradas > 3:
        print("La cantidad de entradas debe ser entre 1 y 3.")
        return
    for _ in range(cantidad_entradas):
        asiento_disponible = False
        while not asiento_disponible:
            mostrar_ubicaciones_disponibles()
            asiento = int(input("Seleccione una ubicación: "))
            asiento_disponible = tipo_entrada.vender_asiento(asiento, input("Ingrese el RUN de la persona: "))

        print("¡Entrada comprada exitosamente!")


def mostrar_ubicaciones_disponibles():
    print("Ubicaciones disponibles:")
    matriz_asientos = []
    for i in range(10):
        fila = []
        for j in range(10):
            asiento = i * 10 + j + 1
            if asiento in platinum.asientos_vendidos:
                fila.append("X")
            else:
                fila.append(str(asiento))
        matriz_asientos.append(fila)

    for fila in matriz_asientos:
        print(" ".join(fila))


def ver_listado_asistentes():
    print("Listado de asistentes:")
    asistentes_ordenados = sorted(asistentes, key=lambda x: x.run, reverse=True)
    for asistente in asistentes_ordenados:
        print(f"RUN: {asistente.run}, Tipo de entrada: {asistente.tipo_entrada}")


def mostrar_ganancias_totales():
    print("Ganancias totales:")
    matriz_ganancias = []
    for entrada in [platinum, gold, silver]:
        cantidad_vendida = len(entrada.asientos_vendidos)
        total = cantidad_vendida * entrada.precio
        fila = [entrada.nombre, str(entrada.precio), str(cantidad_vendida), str(total)]
        matriz_ganancias.append(fila)

    headers = ["Tipo de entrada", "Precio", "Cantidad vendida", "Total"]
    imprimir_matriz(headers, matriz_ganancias)


def seleccionar_tipo_entrada():
    print("Seleccione el tipo de entrada:")
    print("1. Platinum (120,000 CLP)")
    print("2. Gold (80,000 CLP)")
    print("3. Silver (50,000 CLP)")

    opcion = int(input("Ingrese el número de la opción: "))
    if opcion == 1:
        return platinum
    elif opcion == 2:
        return gold
    elif opcion == 3:
        return silver
    else:
        print("Opción inválida.")
        return None


def imprimir_matriz(headers, matriz):
    print("".join(f"{header}\t" for header in headers))
    for fila in matriz:
        print("".join(f"{valor}\t" for valor in fila))


platinum = EntradaConcierto("Platinum", 120000, 20, range(1, 21))
gold = EntradaConcierto("Gold", 80000, 30, range(21, 51))
silver = EntradaConcierto("Silver", 50000, 50, range(51, 101))
asistentes = []

while True:
    mostrar_menu()
    opcion = int(input("Ingrese el número de la opción: "))

    if opcion == 1:
        comprar_entradas()
    elif opcion == 2:
        mostrar_ubicaciones_disponibles()
    elif opcion == 3:
        ver_listado_asistentes()
    elif opcion == 4:
        mostrar_ganancias_totales()
    elif opcion == 5:
        print("¡Gracias por preferir Creativos.cl!")
        break
```
